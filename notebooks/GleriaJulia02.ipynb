```julia
using Plots
using Random
using Colors, ColorSchemes
using DataStructures
```

```julia
EventT = Tuple{Int, Int}

function grow(colonies,space,t,x,y,event_stack,beta)
    function add_event(x,y)
        dt = exp(space[x,y]*beta)+t
        event_stack[(x,y)] = dt
    end
    
    if colonies[x,y]<0
        return false
    end
    if colonies[x,y]==0
        colonies[x,y]=1
        add_event(x+1,y)
        add_event(x-1,y)
        add_event(x,y+1)
        add_event(x,y-1)       
    end
    return true
end;
```

```julia
function run(g_size)
    colonies = fill(Int8(-1),g_size,g_size)
    colonies[2:end-1,2:end-1] .= 0
    space=rand(Float64, (g_size, g_size))
    event_stack = PriorityQueue{EventT, Float64}()
    beta = 25
    grow(colonies,space,0.0,div(g_size,2),div(g_size,2),event_stack,25)
    flag = true
    while(flag)
        (x, y), t = peek(event_stack)
        dequeue!(event_stack)
        flag = grow(colonies, space, t, x, y, event_stack, beta)
    end
    
    colonies
end
```

```julia
function draw(colonies)
    mycmap = ColorScheme([RGB{Float64}(1, 1,1),RGB{Float64}(0, 0,0)])

    g_size = size(colonies)[1]
    heatmap(2:g_size,
        2:g_size, colonies[2:end-1,2:end-1],
        c=cgrad(mycmap),
        xlabel="x", ylabel="y",
        title="Position of colonies in 2d space",
        size=(800,800),
        legend = :none)
end
```

```julia
@time cols = run(10002);
```

```julia
draw(cols)
```
